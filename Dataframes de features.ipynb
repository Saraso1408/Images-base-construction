```python
# função que retorna a norma de um vetor.
def norma(vet):
    norma_vet = la.norm(vet)
    return norma_vet
```

```python
# função que retorna a norma da distancia entre dois vetores, u e v.
def dist_euc(u, v):
    dist = distance.euclidean(u, v, w=None)
    return dist
```

```python
def ang(i1, i2, i): # função que calcula os angulos, dadas as inclinações.
    angle = np.zeros((1924,1))
    a = np.zeros((1924,2))
    
    #for i in range(1924):
    prod = i1[i]*i2[i]
    prod_esc = prod[0]+prod[1]
    s1_norm = norma(i1[i])
    s2_norm = norma(i2[i])
    mult_norm = s1_norm * s2_norm
    a = prod_esc/mult_norm
    angle[i] = math.acos(a)
    
    return angle[i] # Medidas retornadas estão em RADIANOS.
```

```python
# beta = alfa - origem -> como não há lanmarks negativas, considero que a origem está na ponta esquerda da face (superior). 
for i in range(136):
    if i % 2 == 0: # par, ou seja, eixo x
        df.iloc[:, i] = df.iloc[:, i] - df.loc[:, 'land_31_x']
    elif i % 2 != 0: # impar, ou seja, eixo y
        df.iloc[:, i] = df.iloc[:, i] - df.loc[:, 'land_31_y']
df.head()
```

```python
# distancia entre pontos internos dos olhos, para normalização.
d_x = df.loc[:, 'land_43_x'] - df.loc[:, 'land_40_x']
d_y = df.loc[:, 'land_43_y'] - df.loc[:, 'land_40_y']
print(d_x)
print(d_y)
d = (d_x*d_x) + (d_y*d_y)
d = np.sqrt(d)
print("============================\n Distancias de referência para normalização: ", d)
```

```python
# Normalização de todas as coordenadas.
for i in range(136):
    df.iloc[:, i] = df.iloc[:, i]/d
df.head()
#df.shape
```
